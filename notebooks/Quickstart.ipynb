# finanzonline_databox Quickstart

Download documents from the Austrian FinanzOnline DataBox, from the CLI or from Python.

Install the package, then try the CLI and the offline API examples below. `list`, `download` and `sync` all talk to FinanzOnline and need credentials, so this notebook stays with what runs without them.

## 1) Install latest from GitHub (default)

```python
import os
import sys
import sysconfig

# Install uv into this kernel's environment first: a bare `!uv` needs uv on
# PATH, which a reader running this notebook from a plain Python will not have.
# --python then pins the install to the kernel's own interpreter; without it uv
# picks the first environment it discovers and the import below fails.
!{sys.executable} -m pip install --quiet --upgrade uv

!{sys.executable} -m uv pip install --python {sys.executable} --upgrade git+https://github.com/bitranox/finanzonline_databox.git

# Put this kernel's script directory first on PATH, so the `!` cells below run
# the console script that was just installed rather than whichever same-named
# command happens to sit on the ambient PATH.
os.environ["PATH"] = sysconfig.get_path("scripts") + os.pathsep + os.environ["PATH"]
```

## 8) Python API: reading FinanzOnline return codes

Every FinanzOnline call answers with a numeric return code. The domain layer maps it to a meaning, a severity, and whether retrying is worthwhile - no credentials or network involved.

```python
from finanzonline_databox.domain import get_return_code_info

for code in (0, -1, -2):
    info = get_return_code_info(code)
    print(f"code={info.code:<3} severity={info.severity} retryable={info.retryable} :: {info.meaning}")
```
